a) Deslizador que nos cambie o resultado da función:

```python
def f(x):
    f = 3*x+2
    return print(" A función sustituida polo valor de x que lle indica o deslizador da f(", x, ")=", f)
interact(f, x=(-5,10))
```

 d) Producir unha frecuencia  musical

```python
sr = 22050 
T = 2.0   
t = np.linspace(0, T, int(T*sr), endpoint=False) 
x = 0.5*np.sin(2*np.pi*440*t)  
IPython.display.Audio(x, rate=sr)
```
